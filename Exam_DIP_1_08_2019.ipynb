## ข้อ 3 (15 คะแนน) เขียนโปรแกรม Multi-scale Template Matching  

**สิ่งที่ต้องเตรียมตัวก่อนเขียนโปรแกรม**
* ค้นหารูปนักแสดง **Keanu Reeves** จากกลุ่ม DIP-1-2562 และดาวน์โหลดเก็บไว้ในเครื่องคอมฯ
* นำรูปภาพ **Keanu Reeves** ทั้งหมดอัพโหลดไว้ใน **google drive** โดยเก็บไว้ในไดเรกทรอรี่ชื่อ **`Keanu_Reeves`**

**ความสามารถของโปรแกรม**
* อ่านรูปภาพจากไดเร็กทรอรี่ (Directory/Folder)
* แสดงรูปภาพโดยใช้ไลบรารี่ของ matplotlib โดยสามารถกำหนดได้ว่าต้องการแสดงรูปภาพจำนวนกี่คอลัมน์ และแถว (subplot)
* เขียนโปรแกรม **`Multi-scale Template Matching`** เพื่อค้นหาใบหน้าของนักแสดง **Keanu Reeves** จากรูปภาพทุกรูปที่อยู่ในไดเรกทรอรี่
** **`Multi-scale`** หมายถึง สามารถค้นหาได้ทั้งจากรูปภาพที่มีใบหน้าขนาดเล็ก ไปจนถึงใบหน้าที่มีขนาดใหญ่
* แสดงรูปภาพหลังจากที่ทำ Multi-scale Template Matching (รูปภาพที่มีกรอบสี่เหลี่ยมรอบใบหน้านักแสดง) โดยแสดงเป็นแบบ subplot สามารถเลือกจำนวนคอลัมน์ และแถวได้
