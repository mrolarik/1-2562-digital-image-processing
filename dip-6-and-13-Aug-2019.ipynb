```
import matplotlib.pyplot as plt
import random
import numpy as np

for i in range(1, 10):
    plt.subplot(3, 3, i)
    
    rand_num = random.randint(0, 10000)
    show_label = mnist_data.values[rand_num, 0]
    show_img = mnist_data.values[rand_num, 1:]
    show_img = np.reshape(show_img, (28,28))
    rand_flip = random.randint(0,1)
    if(rand_flip == 0):
      show_img = np.fliplr(show_img)
    else:
      show_img = np.flipud(show_img)
    plt.gray()
    plt.title(show_label)
    plt.imshow(show_img)
plt.show()
    
```

```
import matplotlib.pyplot as plt
import random
import numpy as np

fig = plt.figure()
fig.subplots_adjust(hspace=1.2, wspace=0.8, top=0.78)
fig.suptitle('MNIST dataset')
for i in range(1, 7):
    fig.add_subplot(2, 3, i)
    
    rand_num = random.randint(0, 10000)
    show_label = mnist_data.values[rand_num, 0]
    show_img = mnist_data.values[rand_num, 1:]
    show_img = np.reshape(show_img, (28,28))
    #plt.axis('off')
    plt.title(show_label, fontsize=20)
    plt.xlabel('row=' + str(show_img.shape[0]))
    plt.ylabel('col=' + str(show_img.shape[1]))

    plt.gray()
    plt.imshow(show_img)

plt.show()    
```

```
import os

path = '.'

files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
        files.append(os.path.join(r, file))

for f in files:
    print(f)
```

```
import os

path = '.'
files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
        if (('.csv' in file) or ('.jpg' in file)) :
            files.append(os.path.join(r, file))

for f in files:
    print(f)
```
