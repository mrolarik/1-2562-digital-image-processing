# แสดงรูปภาพจากข้อมูล mnist

```
import numpy as np
import matplotlib.pyplot as plt

show_num = 2
mn_df = mnist_df.loc[mnist_df['6'] == show_num]

for i in range(1, 17):
    plt.subplot(4, 4, i)
    
    show_label = mn_df.values[i, 0]
    show_img = mn_df.values[i, 1:]
    show_img = np.reshape(show_img, (28,28))
    plt.gray()
    plt.title(show_label)
    plt.imshow(show_img)
plt.show()    
```

สร้างฟังก์ชัน `show_mnist` เพื่อแสดงรูปภาพ

```
def show_mnist(mn_df):
  for i in range(1, 17):
    plt.subplot(4, 4, i)
    
    show_label = mn_df.values[i, 0]
    show_img = mn_df.values[i, 1:]
    show_img = np.reshape(show_img, (28,28))
    plt.gray()
    plt.title(show_label)
    plt.imshow(show_img)
  plt.show()    
```

```
import numpy as np
import matplotlib.pyplot as plt

show_num = 3
mn_df = mnist_df.loc[mnist_df['6'] == show_num]
show_mnist(mn_df)
```

```
def show_mnist(mn_df, fig_row, fig_col):
  for i in range(1, ((fig_row * fig_col)+1)):
    plt.subplot(fig_row, fig_col, i)
    
    show_label = mn_df.values[i, 0]
    show_img = mn_df.values[i, 1:]
    show_img = np.reshape(show_img, (28,28))
    plt.gray()
    plt.title(show_label)
    plt.imshow(show_img)
  plt.show()  
```

```
show_num = 3
mn_df = mnist_df.loc[mnist_df['6'] == show_num]
fig_row = 3
fig_col = 4
show_mnist(mn_df, fig_row, fig_col)
```

```
for i in range(0,10):
  mn_df = mnist_df.loc[mnist_df['6'] == i]
  show_mnist(mn_df, 2, 3)
```
